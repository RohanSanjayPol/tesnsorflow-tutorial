#### variables in tensorflow

#### 1) creating variables

```python
import tensorflow as tf
intvar=tf.Variable(5)
var1D=tf.Variable([1,2,3])
print(intvar.numpy())
print(var1D.numpy())
```

#### 2) updating variables

```python
var=tf.Variable(20)
# 1) updating variables
var.assign(40)
print(var.numpy())

# 2) add to existing value
var.assign_add(5)
print(var.numpy())

# 3) sub to existing value
var.assign_sub(2)
print(var.numpy())
```

#### 3) variables in neural network

```python
weights=tf.Variable(tf.random.normal([3,2]))
bias=tf.Variable(tf.zeros([2]))
print(weights.numpy())
print(bias.numpy())
```

#### 4). Why Variables are Important?

Store trainable parameters of ML/DL models.

Allow gradient updates (via backpropagation).

Work seamlessly with tf.GradientTape for automatic differentiation.

#### 5) changing the shape of variable

```python
var=tf.Variable([1,2,3,4,5,6,7,8,9])
print(var.numpy())
# change the shape
tf.reshape(var,[3,3])
```

#### 6) changing the data type of tensor

```python
var=tf.Variable([1,2,3,4,5],dtype=float)
print(var.numpy())
```

#### 7) operations with variables

```python
var1=tf.Variable([1,2])
var2=tf.Variable([2,2])
# addition operation
add=tf.add(var1,var2)
print(add.numpy())
# substraction operation
sub=(var1-var2)
print(sub.numpy())
```

#### 8) hardware selection for variables

```python
var=tf.Variable(20)
print(var.device)
```
