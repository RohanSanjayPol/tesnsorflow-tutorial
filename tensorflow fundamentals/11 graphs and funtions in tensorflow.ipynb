#### 1. What is a Graph in TensorFlow?

#### 2. Functions in TensorFlow (@tf.function)

The @tf.function decorator converts a Python function into a TensorFlow graph automatically.

Inside the graph, TensorFlow executes faster and is more efficient.

#### 3. Example (Normal vs Graph Mode)

```python
import tensorflow as tf

# normal function
def simple_add(a,b):
    return a+b

# graph mode
@tf.function
def graph_add(a,b):
    print(a+b)
    
a=tf.constant(2)
b=tf.constant(3)

print("eager (python):",simple_add)
print("Graph (tf.funtion)",graph_add)
    
```

#### 4. Example: Graph with Control Flow

```python
@tf.function
def my_funt(x):
    if x>0:
        return 'positive'
    else:
        return 'negative'
    
print(my_funt(tf.constant(3)).numpy())
print(my_funt(tf.constant(-2)).numpy())
```
