#### 1. Basic Arithmetic

```python
import tensorflow as tf
a=tf.constant([1,2,3],dtype=float)
b=tf.constant([2,3,5],dtype=float)
# addition
add=tf.add(a,b)
print('additon',add.numpy())
sub=tf.subtract(a,b)
print('subtract',sub.numpy())
mul=tf.multiply(a,b)
print('multiply',mul.numpy())
div=tf.divide(a,b)
print('divide',div.numpy())
```

#### 2. Power & Roots

```python
a=tf.constant([2,3,5,16])
# power
power=tf.pow(a,2)
print(power.numpy())
# square root
#square_root=tf.sqrt(a)
#print(square_root)
```

#### 3. Exponentials & Logarithms

```python
a=tf.constant([1,2,3],dtype=float)
# exponentials
exp=tf.exp(a)
sigmoid_activation=1/(1+tf.exp(a))
print(sigmoid_activation.numpy())
# logarithmic
log=tf.math.log(a)
print(log.numpy())
```

#### 4. Rounding & Absolute

```python
c=tf.constant([-1.1,2.6,3.9],dtype=float)
# abs
absoulate=tf.abs(c)
print(absoulate.numpy())
# round
Round=tf.round(c)
print(Round.numpy())
# floor
floor=tf.floor(c)
print(floor.numpy())
# ceil
ceil=tf.math.ceil(c)
print(ceil.numpy())
```

#### 5. Trigonometric Functions

```python
x = tf.constant([0.0, 3.14159/2])
print(tf.sin(x).numpy())
print(tf.cos(x).numpy())
print(tf.tan(x).numpy())
```

#### 6. Reduction Operations

```python
matrix = tf.constant([[1, 2, 3], [4, 5, 6]])
print('sum',tf.reduce_sum(matrix).numpy())
print('mean',tf.reduce_mean(matrix).numpy())
print('max',tf.reduce_max(matrix).numpy())
print('min',tf.reduce_min(matrix).numpy())
print('accross rows: ',tf.reduce_sum(matrix,axis=1).numpy())
```

#### 7. Linear Algebra (tf.linalg)

```python
M = tf.constant([[1., 2.], [3., 4.]])
# determinant
determinant=tf.linalg.det(M)
print('determinant',determinant.numpy())
# inverse
inverse=tf.linalg.inv(M)
print('inverse: ',inverse.numpy())
# matrix multiplication
mat_mul=tf.linalg.matmul(M,M)
print('matrix multiplication: ',mat_mul.numpy())
```

#### 8. Useful Special Ops

```python
a=tf.constant([2,4,6],dtype=float)
# square
print(tf.math.square(a).numpy())
# standard deviation
print(tf.math.reduce_std(a).numpy())
# variace
print(tf.math.reduce_variance(a).numpy())
```
