#### string tensors

```python
import tensorflow as tf
string_tensor1=tf.constant(['hello rohan'],dtype=tf.string)
print(string_tensor1)
```

```python
string_tensor2=tf.constant('kaise ho',dtype=tf.string)
print(string_tensor2)
```

#### string operations

###### 1) tf.strings.length()

```python
tf.strings.length(['cat','dog','deep learning'])
```

###### 2) tf.strings.split()

```python
tf.strings.split(['hello how can i be an expert in Deep learning'])
```

###### 3) tf.strings.join()

```python
tf.strings.join(['hello',' ','good night','everyone','and','how' 'are' 'you'])
```

###### 4) tf.strings.nicode_decode

```python
tf.strings.unicode_decode('hello','utf8')
```

###### 5) tf.strings.unicode_encode()

```python
decoded_tensor=tf.ragged.constant([22,600])
tf.strings.unicode_encode(decoded_tensor,"UTF-8")
```

###### 6) tf.equal()

```python
string1=tf.constant(['hello'],dtype=tf.string)
string2=tf.constant(['i am rohan'],dtype=tf.string)
print(tf.equal(string1,string2))
```

###### 7) pattern matching
 
tf.strings.regex_full_match()

```python
pat1=tf.constant("Ten")
pat2=tf.constant("ten")
print(tf.strings.regex_full_match(pat1,pat2))
```

###### 8) tf.strings.lower

```python
upper_case=tf.constant(['I Love Tensorflow But still I go with pytorch'])
print(tf.strings.lower(upper_case))
```
