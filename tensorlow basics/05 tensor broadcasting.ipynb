### 1) different shapes

```python
import tensorflow as tf
t1=tf.constant([[1,2,3],
                [4,5,6]])
print(t1)
print(tf.rank(t1).numpy())
```

```python
t2=tf.constant([[11,12,13]])
print(t2)
print(tf.rank(t2).numpy())
```

```python
# performing the arithmetic operations and its ranks are same
# addition operation
t3=t1+t2
print(t3)
# substraction operation
t4=t1-t2
print(t4)
```

### 2) different shapes

```python
# 2D matrix
t1=tf.constant([[1,2,3],
                [4,5,6]])
print(t1)
print(tf.rank(t1).numpy())
# 1D vevtor
t2=tf.constant([1,2,3])
print(t2)
print(tf.rank(t2).numpy())
```

```python
# Arithmetic operations
t3=t1+t2
print(t3)
```

```python
t1=tf.constant([[1,2,3],
               [4,5,6]])
print(t1)
print(tf.rank(t1).numpy())
t2=tf.constant([2])
print(t2)
print(tf.rank(t2).numpy())
```

```python
# Arithmetic operations
t3=t1+t2
print(t3)
```

```python
# 3D tensor
t1=tf.constant([[[1,2,3],
                [4,5,6],
                [7,8,9]],
               
                [[2,2,2],
                 [3,3,3],
                 [4,4,4]]])
print(t1)
print(tf.rank(t1).numpy())
```

```python
# 2D tensor
t2=tf.constant([[1,2,3],
                [4,5,6],
                [7,8,9]])
print(t2)
print(tf.rank(t2).numpy())
```

```python
# adding new axis to t2
broadcasted_t2=tf.expand_dims(t2,axis=-1)
broadcasted_t2
```

```python
t3=t1+t2
print(t3)
```
