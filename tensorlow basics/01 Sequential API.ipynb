## Sequential API

```python
#importing the necessary libaraires
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
# building a sequential model
model=Sequential([
    Flatten(input_shape=(28,28)),#converts 28*28 input into 784 vector
    Dense(128,activation='relu'),#hidden layer
    Dense(10,activation='softmax')#output layer
])
```

```python
# compile the model
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# train the model

#load the dataset
from tensorflow.keras.datasets import mnist
#split the dataset
(x_train,y_train),(x_test,y_test)=mnist.load_data()
#normalize the range between 0 to 1
x_train,x_test=x_train/255.0,x_test/255.0
# train the model
model.fit(x_train,y_train,epochs=15,validation_data=(x_test,y_test))
```

```python
# evaluate the model
test_loss,test_acc=model.evaluate(x_test,y_test)
print('Test Accuracy',test_acc)
print('test loss',test_loss)
```

```python
# make predications
predictions=model.predict(x_test[:5]) #probablties of class
print(predictions)
print(predictions.argmax(axis=1))# final predicted class
print(y_test[:5])#actual labels
```
