#### tf.reshape(tensors,shape,name=none)

```python
import tensorflow as tf
t1=tf.constant([[1,2,3],
    [4,5,6],
    [7,8,9]])
print(t1)
print('shape of the tensors: ',tf.shape(t1).numpy())
```

```python
# reshaping it into 1D tensors( vectors)
reshaped_t1=tf.reshape(t1,[9])
print(reshaped_t1)
```

```python
t2=tf.constant([1,2,3,4,5,6,7,8,9])
print(t2)
```

```python
# reshaping it into 2D tensors(matrix)
reshaped_t2=tf.reshape(t2,[3,3])
print(reshaped_t2)
```

#### transposition operation to reshape

```python
t3=tf.constant([[1,2,3,4],
                [5,6,7,8]
])
print(t3)
```

```python
transpose_t3=tf.transpose(t3,perm=[1,0])
transpose_t3
```
