```python
import tensorflow as tf
sparse=tf.sparse.SparseTensor(
 indices=[[0,2],[2,0],[2,2]],
 values=[3,4,5],
 dense_shape=[3,3]    
)
print(sparse)
```

###### tf.sparse.to_dense()

```python
# converting back sparse to dense
tf.sparse.to_dense(sparse)
```

###### tf.sparse.from_dense()

```python
dense=tf.constant([[0,0,1],
                   [0,3,4],
                   [0,0,5]])
print(dense)
```

```python
sparse_tensor=tf.sparse.from_dense(dense)
sparse_tensor
```

#### Arithmetic manipulations

```python
sparse_tensor1=tf.sparse.SparseTensor(
indices=[[0,2],[1,1],[1,2],[2,0]],
values=[1,2,3,4],
dense_shape=[3,3]    
)
sparse_tensor2=tf.sparse.SparseTensor(
indices=[[1,1],[1,2],[2,1],[2,0]],
values=[1,2,3,4],
dense_shape=[3,3]    
)
print(sparse_tensor1)
print(sparse_tensor2)
```

```python
sparse_tensor3=tf.sparse.add(sparse_tensor1,sparse_tensor2)
print(sparse_tensor3)
```

#### sparse concatination

```python
tf.sparse.concat(axis=0,sp_inputs=[sparse_tensor1,sparse_tensor2])
```
