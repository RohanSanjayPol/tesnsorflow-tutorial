```python
import tensorflow as tf
```

```python
# tensors shape
tensors=tf.constant([[1,2,3],[45,34,26]])
print('tensors',tensors)
shape=tensors.shape
print('tesors shape',shape)
```

```python
# size of tensors
Size=tf.size(tensors)
print('size of tensors',Size)
```

### rank of tensors

```python
# create tensors of different ranks
scaler=tf.constant(1)
vector=tf.constant([1,2])
matrix=tf.constant([[2,3,4],[45,74,34]])
tensor_3d=tf.constant([[[1,2],[3,4]],[[45,65],[34,67]]])

# get the rank of tensors
rank_scaler=tf.rank(scaler)
rank_vector=tf.rank(vector)
rank_matrix=tf.rank(matrix)
rank_tensor_3d=tf.rank(tensor_3d)

# print the ranks of tensors
print('rank of scaler: ',rank_scaler.numpy())
print('rank of vector: ',rank_vector.numpy())
print('rank of matrix: ',rank_matrix.numpy())
print('rank of 3d tensors: ', rank_tensor_3d.numpy())
```
