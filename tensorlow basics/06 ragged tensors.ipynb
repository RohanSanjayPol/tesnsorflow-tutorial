###### tf.ragged.constant()

```python
import tensorflow as tf
tensors=tf.ragged.constant([[1,2,3],
                            [4,5],
                            [6]])
print(tensors)
```

```python
# operations we can do
print(tensors[0])
print(tensors[1,1])
#convet ragged to dense
print(tensors.to_tensor(default_value=0))
```

###### tf.RaggedTensors.from_value_rowids

```python
words=[1,2,3,4,5,6,7,8,9]
words_ids=[0,0,0,0,1,1,2,2,2]
tf.RaggedTensor.from_value_rowids(values=words,value_rowids=words_ids)
```
