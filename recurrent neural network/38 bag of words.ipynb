```python
# importing the necessary libararies
import tensorflow as tf
```

```python
# define sample data
text_data = [
    'i am rohan rohan rohan pol pol',
    'i will do'
]
print(text_data)
```

```python
# create and configuration TextVectorization layer
max_tokens=20
output_mode='count'
vectorization=tf.keras.layers.TextVectorization(
    max_tokens=max_tokens,
    output_mode=output_mode
)
vectorization.adapt(text_data)
```

```python
bow_representation=vectorization(text_data)
print('vocabulary',vectorization.get_vocabulary())
print('bag of word representation',bow_representation.numpy())
```
