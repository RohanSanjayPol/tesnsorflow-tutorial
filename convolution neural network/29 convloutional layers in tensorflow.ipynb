```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Conv1D,Conv2D,Conv3D
```

#### 1. Conv1D

```python
model=Sequential([
    Conv1D(filters=64,kernel_size=2,strides=1,padding='valid',activation='relu')
])
```

#### 2.Conv2D

```python
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',input_shape=(28,28,3)))
```

```python
model.summary()
```

#### 3.Conv3D

```python
model2=Sequential()
model2.add(Conv3D(filters=64,kernel_size=(3,3,3),strides=(1,1,1),padding='valid',activation='relu',use_bias=True,input_shape=(28,28,3,4)))
```

### Depthwise convolutional layers

#### 2. DepthwiseConv2D

```python
from tensorflow.keras.layers import DepthwiseConv2D
```

```python
model=Sequential()
model.add(DepthwiseConv2D(kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',input_shape=(28,28,3)))
model.summary()
```

### pooling layers in tensorflow

#### 1. MaxPooling2D

```python
from tensorflow.keras.layers import MaxPooling2D,AveragePooling2D,GlobalAveragePooling2D
```

```python
model=Sequential()
model.add(MaxPooling2D((2,2),padding='valid',input_shape=(28,28,3)))
model.summary()
```

#### 2. AveragePooling

```python
model.add(AveragePooling2D(pool_size=(2,2),padding='valid',strides=None))
model.summary()
```
