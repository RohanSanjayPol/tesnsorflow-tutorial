```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dense,Conv2D,MaxPooling2D,Dropout,LeakyReLU
```

#### 1. relu activation

```python
model=Sequential([
    Dense(128,activation='relu',input_shape=(11,)),
    Dense(1,activation='softmax')
])
model.summary()
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

#### 2. sigmoid activation

```python
model=Sequential([
    Dense(128,activation='sigmoid',input_shape=(3,)),
    Dense(1,activation='sigmoid')
])
model.summary()
model.compile(optimizer='Adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

#### 3. tanH (Hyperbolic Tangent)

```python
model=Sequential([
    Dense(128,activation='tanh',input_shape=(3,)),
    Dense(5,activation='softmax')
])
model.summary()
model.compile(optimizer='adam',
             loss='categrorical_crossentopy',
             metrics=['accuracy'])
```

#### 4. Softmax

```python
model=Sequential([
    Dense(128,activation='softmax',input_shape=(12,))
])
model.summary()
```

#### 5. Leaky ReLU

```python

model=Sequential([
    Dense(128,input_shape=(11,)),
    LeakyReLU(alpha=0.3)
])
model.summary()
```

#### 6. ELU activation

```python
model=Sequential([
    Dense(128,activation='elu',input_shape=(11,))
])
model.summary()
```

#### 7. swish activation

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='swish', input_shape=(32,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```
