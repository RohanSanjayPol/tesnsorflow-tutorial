```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,LSTM,Dropout,BatchNormalization
```

#### 1. Dense layer

```python
model=Sequential([
    Dense('128',activation='relu')
])
model.summary()
```

#### 2. Conv2D layer

```python
model=Sequential([
    Conv2D(filters=32,kernel_size=(3,3),activation='relu')
])
model.summary()
```

#### 3. pooling layer

```python
model=Sequential([
    MaxPooling2D(pool_size=(2,2))
])
model.summary()
```

#### 4. Recurrent Layer


```python
model=Sequential([
    LSTM(units=128)
])
model.summary()
```

#### 5. Dropout layer

```python
model=Sequential([
    Dropout(rate=0.5)
])
model.summary()
```

#### 6. Batch Normalization Layer

```python
model=Sequential([
    BatchNormalization()
])
model.summary()
```
