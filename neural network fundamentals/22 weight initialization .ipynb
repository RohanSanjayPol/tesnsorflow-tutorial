#### 2. weight initialization using He Normal Initialization

```python
import tensorflow as tf
def he_weight_initialization(fan_in,fan_out):
    limit=tf.sqrt(2/fan_in)
    return tf.random.uniform([fan_in,fan_out],minval=-limit,maxval=limit,dtype=tf.float32)
```

```python
fan_in=4
fan_out=5
w1=he_weight_initialization(fan_in,fan_out)
print(w1.numpy())
b1=tf.zeros(fan_out)
print(b1.numpy())
```
